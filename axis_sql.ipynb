```python
import pandas as pd
import os
from sqlalchemy import create_engine
```

```python

base_path = r"C:\Users\HP\OneDrive\Documents\axis_bank_recommendation\data\sql_upload"

print(os.listdir(base_path))

#  Read files correctly
customers_df = pd.read_csv(f"{base_path}/customers.csv")
accounts_df = pd.read_csv(f"{base_path}/accounts.csv")
transactions_df = pd.read_csv(f"{base_path}/transactions.csv")
features_df = pd.read_csv(f"{base_path}/feature_store.csv")
product_catalog_df = pd.read_csv(f"{base_path}/product_catalog.csv")
recommendations_df = pd.read_csv(f"{base_path}/recommendations.csv")
model_metadata_df = pd.read_csv(f"{base_path}/model_metadata.csv")
```

```python

username = "postgres"
password = "Kovilvenni"
host = "localhost"
port = "5432"
database = "axis_bank_ml"

engine = create_engine(
    f"postgresql://{username}:{password}@{host}:{port}/{database}"
)

print("Database connected successfully")
```

```python
customers_df.to_sql("customers", engine, if_exists="replace", index=False)

accounts_df.to_sql("accounts", engine, if_exists="replace", index=False)

transactions_df.to_sql("transactions", engine, if_exists="replace", index=False)

features_df.to_sql("feature_store", engine, if_exists="replace", index=False)

product_catalog_df.to_sql("product_catalog", engine, if_exists="replace", index=False)

recommendations_df.to_sql("recommendations", engine, if_exists="replace", index=False)

model_metadata_df.to_sql("model_metadata", engine, if_exists="replace", index=False)

print("Tables uploaded successfully")
```
